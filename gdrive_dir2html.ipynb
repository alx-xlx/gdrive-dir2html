```python
#@title <- Run this Code and Enter the Path of Folder to Index
import os
import sys
import datetime
from os.path import getsize
from IPython.display import clear_output
from time import sleep
from colorama import Fore
from time import gmtime, strftime

# global variables definition
appName = "gdrive-dir2html"
appVer = "1.0"
genDate = datetime.datetime.now().strftime("%d/%m/%Y")
genTime = datetime.datetime.now().strftime("%H:%M")
appLink = "https://github.com/alx-xlx/gdrive-dir2html"
dirData = ""
numFiles=0 
numDirs=0 
grandTotalSize=0
linkFiles="false" # file linking not yet implemented

folder_to_index = ""
global TotalFiles
global TotalFolders


class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   RESET = '\033[0m'

# functions definition
def generateDirArray(dirToScan):
    global dirData
    global numFiles
    global numDirs
    global grandTotalSize

    # assing a number identifier to each directory
    i = 1
    dirIDsDictionary = {}
    dirIDsDictionary[dirToScan] = 0

    lengthDir = 0
    lengthFiles = 0

    # lengthCD = 0

    # for currentDir, dirs, files in os.walk(dirToScan):
    #     clear_output(wait=True)
    #     lengthDir = lengthDir + len(dirs)
    #     lengthFiles = lengthFiles + len(files)
    #     lengthCD = lengthCD + len(currentDir)
    #     print(dirs)
    #     print('----------------------------------')
    #     print(files)
    #     print('----------------------------------')
    #     print(currentDir)
    #     print(str(lengthDir) + ' | ' + str(lengthFiles) + ' | ' + str(lengthCD))
    #     sleep(10)
    # print('DONE')
    # sleep(5)


    for currentDir, dirs, files in os.walk(dirToScan):
        clear_output(wait=True)
        lengthDir = lengthDir + len(dirs)
        lengthFiles = lengthFiles + len(files)
        print(Fore.YELLOW + 'Phase I' + Fore.BLUE + ' - (Assigning Directories with Identifiers) : \n' + color.RESET)
        print(Fore.RED + color.BOLD + '\nFolders : ' + color.RESET + Fore.RED + str(lengthDir))
        print(Fore.LIGHTGREEN_EX + color.BOLD + 'Files : ' + color.RESET + Fore.GREEN + str(lengthFiles))
        j = 0
        print(Fore.BLUE + color.BOLD + currentDir.replace('/content/drive/Shared drives/', '').replace('/', '\n')+color.RESET)
        # print(currentDir.replace('/','\n'))
        for dir in dirs:
            # clear_output(wait=True)
            j = j + 1
            
            print(Fore.LIGHTBLACK_EX + str(j)+'/'+str(len(dirs)) + ' - ' + Fore.RED + color.BOLD + dir + color.RESET)
            dirIDsDictionary[currentDir+'/'+dir] = i
            i = i + 1
    TotalFiles = lengthFiles
    TotalFolders = lengthDir
          
    print(Fore.LIGHTGREEN_EX + 'Phase I - Folder Scan Completed | Total Folders Scanned - ' + str(i))
    # sleep(2)

    # initilize array to hold all dir data, dimensioning it to hold the total number of dirs
    allDirArray=[]
    for p in range(i):
        allDirArray.append(p)

    file_count = 0
    dir_count = 0
    lengthDir = 0
    lengthFiles = 0
    # traverse the directory tree     
    for currentDir, dirs, files in os.walk(dirToScan):
        # clear_output(wait=True)
        # count = count + 1
        # print('Current Folder', currentDir)
        currentDirId = dirIDsDictionary[currentDir]
        currentDirArray = []  # array to hold all current dir data
        currentDirModifiedTime = datetime.datetime.fromtimestamp(os.path.getmtime(currentDir))
        currentDirModifiedTime = currentDirModifiedTime.strftime("%d/%m/%Y %H:%M:%S")
        currentDirFixed = currentDir.replace("/","\\\\")    # replace / with \\ in the dir path (necessary for javascript functions to work properly
        currentDirArray.append(currentDirFixed+'*0*'+currentDirModifiedTime)   # append directory info to currentDirArray 
        totalSize = 0
        f = 0
        # print()
        for file in files:
            clear_output(wait=True)
            # lengthDir = lengthDir + len(dirs)
            # lengthFiles = lengthFiles + len(files)
            file_count = file_count + 1
            print(Fore.YELLOW + 'Phase II\n')
            print(Fore.RED + color.BOLD + 'Directory Scanning : '+ color.RESET+Fore.YELLOW + str(dir_count)+Fore.WHITE+'/'+Fore.RED+str(TotalFolders))
            print(Fore.LIGHTGREEN_EX + color.BOLD + 'Files Scanning : '+ color.RESET+Fore.CYAN + str(file_count)+Fore.WHITE+'/'+Fore.GREEN+str(TotalFiles))
            print('\n')
            print(Fore.BLUE + color.BOLD + currentDir.replace('/content/drive/Shared drives/', '').replace('/', '\n'))
            print('\n')
            # print(str(f)+'/'+str(len(files)) + ' - ' + file)
            try:
              print(Fore.LIGHTBLACK_EX + str(dir_count)+'/'+str(TotalFolders) + ' : '+Fore.RED + dir)
            except:
              pass
            print(Fore.LIGHTBLACK_EX + str(file_count)+'/'+str(TotalFiles) + ' : '+Fore.GREEN + file)
            numFiles = numFiles + 1
            fileSize = getsize(currentDir+'/'+file)
            totalSize = totalSize + fileSize
            grandTotalSize = grandTotalSize + fileSize
            fileModifiedTime = datetime.datetime.fromtimestamp(os.path.getmtime(currentDir+'/'+file))
            fileModifiedTime = fileModifiedTime.strftime("%d/%m/%Y %H:%M:%S")
            currentDirArray.append(file+'*'+str(fileSize)+'*'+fileModifiedTime)   # append file info to currentDirArray
        # print()

        currentDirArray.append(totalSize)   # append total file size to currentDirArray
        # create the list of directory IDs correspondent to the subdirs present on the current directory
        # this acts as a list of links to the subdirectories on the javascript code
        dirLinks = ''
        # d_count = 0
        dir = ''
        for dir in dirs:
            clear_output(wait=True)
            lengthDir = lengthDir + len(dirs)
            lengthFiles = lengthFiles + len(files)
            dir_count = dir_count + 1
            print(Fore.YELLOW + 'Phase II\n')
            print(Fore.RED + color.BOLD + 'Directory Scanning : '+ color.RESET+Fore.YELLOW + str(dir_count)+Fore.WHITE+'/'+Fore.RED+str(TotalFolders))
            print(Fore.LIGHTGREEN_EX + color.BOLD + 'Files Scanning : '+ color.RESET+Fore.CYAN + str(file_count)+Fore.WHITE+'/'+Fore.GREEN+str(TotalFiles))
            print('\n')
            print(Fore.BLUE + color.BOLD + currentDir.replace('/content/drive/Shared drives/', '').replace('/', '\n'))
            print('\n')
            print(Fore.LIGHTBLACK_EX + str(dir_count)+'/'+str(TotalFolders) + ' : '+Fore.RED + dir)
            try:
              print(Fore.LIGHTBLACK_EX + str(file_count)+'/'+str(TotalFiles) + ' : '+Fore.GREEN + file)
            except:
              pass
            numDirs = numDirs + 1
            dirLinks = dirLinks + str(dirIDsDictionary[currentDir+'/'+dir]) + '*'
        # print('Files - ' + str(f)+'/'+str(len(files)) + ' | Folders - ' + str(d_count)+'/'+str(len(dirs)))
        
        dirLinks = dirLinks[:-1]    # remove last *
        currentDirArray.append(dirLinks)
        allDirArray[currentDirId]=currentDirArray   # store currentDirArray on the correspondent position of allDirArray
        # os.system('clear')
        # output.clear()
        # clear_output(wait=True)
    # gTotalFiles = lengthFiles
    # gTotalFolders = lengthFolders
    # print(gTotalFiles)
    # print(gTotalFolders)
    # sleep(5)
    # from allDirArray, generate the text to replace [DIR DATA] on HTML file
    #
    # dirData format:
    #
    #   dirs[DIRECTORY_ID] =
    #   "DIRECTORY_PATH*0*MODIFIED_TIME",
    #   "FILENAME*FILESIZE*MODIFIED_TIME",
    #   ...
    #   TOTAL_FILE_SIZE,
    #   "SUBDIRECTORY_ID*SUBDIRECTORY_ID*SUBDIRECTORY_ID*...",
    #   ];
    print(Fore.YELLOW + 'Scanning Folder Tree Completed !!')
    
    list_data = []
    for d in range(len(allDirArray)):
        # print('allDirArray')

        list_data.append("dirs[" + str(d) + "] = [\n")
        for g in range(len(allDirArray[d])):
            if type(allDirArray[d][g]) == int:
                list_data.append(str(allDirArray[d][g])+",\n")
            else:
                list_data.append('"'+allDirArray[d][g]+'",\n')
        list_data.append("];\n")
        list_data.append("\n")
    dirData += ''.join(list_data)

    print(Fore.YELLOW + 'Adding Data Over')
    return 



def generateHTML(dirData,appName,appVer,genDate,genTime,title,appLink,numFiles,numDirs,grandTotalSize,linkFiles):
    print('Generating HTML file')
    templateFile = open('gdrive-dir2html/example/template.html', 'r')
    indexGenTime = strftime("%d-%m-%Y-%H-%M-%S", gmtime())
    indexName = title+'-'+indexGenTime+'.html'
    outputFile = open(title+'-'+indexGenTime+'.html', 'w')
    op = 0
    for line in templateFile:
        clear_output(wait=True)
        op = op + 1
        print(Fore.LIGHTGREEN_EX+'Generating : '+Fore.RED+str(indexName))
        print(Fore.GREEN + str(op))
        modifiedLine = line
        modifiedLine = modifiedLine.replace('[DIR DATA]', dirData)
        modifiedLine = modifiedLine.replace('[APP NAME]', appName)
        modifiedLine = modifiedLine.replace('[APP VER]', appVer)
        modifiedLine = modifiedLine.replace('[GEN DATE]', genDate)
        modifiedLine = modifiedLine.replace('[GEN TIME]', genTime)
        modifiedLine = modifiedLine.replace('[TITLE]', title)
        modifiedLine = modifiedLine.replace('[APP LINK]', appLink)
        modifiedLine = modifiedLine.replace('[NUM FILES]', str(numFiles))
        modifiedLine = modifiedLine.replace('[NUM DIRS]', str(numDirs))
        modifiedLine = modifiedLine.replace('[TOT SIZE]', str(grandTotalSize))
        modifiedLine = modifiedLine.replace('[LINK FILES]', linkFiles)
        outputFile.write(modifiedLine)
    templateFile.close()
    outputFile.close()
    clear_output(wait=True)
    raw_size = int(grandTotalSize)
    # raw_size = raw_size/1024
    if raw_size/1024 < 1024:
      a = "%.2f" % (raw_size/1024)
      size = str(a) + ' KB'
    elif raw_size/1048576 < 1024:
      a = "%.2f" % (raw_size/1048576)
      size = str(a) + ' MB'
    else:
      a = "%.2f" % (raw_size/(1024*1024*1024))
      size = str(a) + ' GB'

      

    print(Fore.BLUE + 'Total Files : '+str(numFiles))
    print(Fore.YELLOW + 'Total Folders : '+str(numDirs))
    print(Fore.RED + 'Total Size : '+Fore.GREEN +str(size))
    print(Fore.LIGHTGREEN_EX + 'Generated : '+ Fore.RED + str(indexName))
    




# main program start point
if len(sys.argv) < 1:   # check if required arguments are supplied
    print("Missing arguments. This tool should be used as follows:")
    print ("gdrive-dir2html.py pathToIndex outputFileName")
else:
    # folder_to_index = '/content/drive/Shared drives/Tutorials | Product of Web Premium. Pw/2. C# Intermediate Classes, Interfaces and OOP'
    folder_to_index = input("Copy & Paste : path of Folder you want to Index : ")
    print("Processing....")
    pathToIndex = str(folder_to_index)
    # title = str(sys.argv[2])
    title = str('/content/index')
    if os.path.exists(pathToIndex): # check if the specified directory exists
        generateDirArray(pathToIndex)
        generateHTML(dirData,appName,appVer,genDate,genTime,title,appLink,numFiles,numDirs,grandTotalSize,linkFiles)
    else:
        print("The specified directory doesn't exist")
```
